````python
word = "apple" #this line is used to state wich word is chosen
count = 0 #this starts the count of the repeated occurances
with open("temp.txt", 'r') as f: #This line opens a txt file in read mode which contains all the words and states its name is f.
    for line in f: #This checks each line in f
        words = line.split() #this checks each word in each line
        for i in words: #this is what checks if words are a match and adds to the count.
            if(i==word): 
                count=count+1
print("Occurrences of the word", word, ":", count)#this outputs the final result of the count.
````

A VPN connection establishes a secure connection between you and the internet. Via the VPN, all your data traffic is routed through an encrypted virtual tunnel. This disguises your IP address when you use the internet, making its location invisible to everyone. A VPN connection is also secure against external attacks.
Pros of VPN:
    Secures your browsing data
    Improves your daily cybersecurity

Cons of VPN:
    Slows down your internet speed

Unlike symmetric encryption, which uses the same secret key to encrypt and decrypt sensitive information, asymmetric encryption, also known as public-key cryptography or public-key encryption, uses mathematically linked public- and private-key pairs to encrypt and decrypt senders' and recipients' sensitive data.
Example, If one key would be leaked in symmetric encryption then the whole process is null. However in Assymetric encryption the secret key would never be leaked.
