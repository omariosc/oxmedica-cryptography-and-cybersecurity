```python

import sqlite3

# Connect to SQLite database (create one if not exists)
conn = sqlite3.connect('library.db')
cursor = conn.cursor()

# Create Books table
cursor.execute('''
    CREATE TABLE IF NOT EXISTS Books (
        book_id INTEGER PRIMARY KEY,
        title TEXT,
        author TEXT,
        isbn TEXT
    )
''')

# Create Members table
cursor.execute('''
    CREATE TABLE IF NOT EXISTS Members (
        member_id INTEGER PRIMARY KEY,
        name TEXT,
        email TEXT,
        address TEXT
    )
''')

# Create Borrowing table
cursor.execute('''
    CREATE TABLE IF NOT EXISTS Borrowing (
        borrowing_id INTEGER PRIMARY KEY,
        member_id INTEGER,
        book_id INTEGER,
        borrow_date DATE,
        return_date DATE,
        FOREIGN KEY (member_id) REFERENCES Members(member_id),
        FOREIGN KEY (book_id) REFERENCES Books(book_id)
    )
''')

# Commit changes and close connection
conn.commit()
conn.close()

print("Tables created successfully.")

# Reconnect to the database
conn = sqlite3.connect('library.db')
cursor = conn.cursor()

# Add a new book
cursor.execute('''
    INSERT INTO Books (title, author, isbn)
    VALUES ('The Great Gatsby', 'F. Scott Fitzgerald', '9780141182636')
''')

# Add a new member
cursor.execute('''
    INSERT INTO Members (name, email, address)
    VALUES ('John Doe', 'johndoe@example.com', '123 Main St, Anytown')
''')

# Add a new borrowing
cursor.execute('''
    INSERT INTO Borrowing (member_id, book_id, borrow_date, return_date)
    VALUES (1, 1, '2024-07-02', NULL)
''')

# Commit changes and close connection
conn.commit()
conn.close()

print("Data added successfully.")

# Simulated SQL injection (not to be used in real applications)
unsafe_input = "1 OR 1=1; --"
cursor.execute(f'''
    SELECT * FROM Members WHERE member_id = {unsafe_input}
''')

# Fetch all data
rows = cursor.fetchall()
for row in rows:
    print(row)

# Using parameterized query to prevent SQL injection
safe_input = (1,)
cursor.execute('''
    SELECT * FROM Members WHERE member_id = ?
''', safe_input)

# Fetch all data
rows = cursor.fetchall()
for row in rows:
    print(row)

```
