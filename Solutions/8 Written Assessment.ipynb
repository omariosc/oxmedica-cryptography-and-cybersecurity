```python
def counting(list, word):
    count = 0
    for i in range(0, len(list)):

        if (word == list[i]):
           count = count + 1

    return count


user_input = input("Enter Text: ")
list = user_input.split()

number = []
for i in range(0, len(list)):
    count = counting(list, list[i])
    number.insert(i, count)

for i in range(0, len(list)):
    print (f"{list[i]} is repeated {number[i]} times.")
```
